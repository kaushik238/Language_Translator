```python
# Required imports
from tkinter import *
from tkinter import ttk, messagebox
from deep_translator import GoogleTranslator

# Get the list of all supported languages from Google Translator
languages = GoogleTranslator().get_supported_languages(as_dict=True)

# Create a language list from the values (language names)
language_list = list(languages.values())

# Create the main application window
root = Tk()
root.geometry("700x400")
root.title("Language Translator")
root.config(bg='#D3D3D3')

# Title Label
Label(root, text="Language Translator", font=('Arial', 20), bg='#8B8B7A', fg='Black', width=50, pady=5).pack(pady=10)

# Creating frames for text areas
f1 = Frame(root, width=320, height=200, bg='white', bd=5)
f1.place(x=15, y=70)
f2 = Frame(root, width=320, height=200, bg='white', bd=5)
f2.place(x=360, y=70)

# Translation function
def trans_late():
    txt2.delete(1.0, END)  # Clear previous output
    try:
        # Identify the language keys based on user selection
        from_language_key = list(languages.keys())[list(languages.values()).index(c1.get())]
        to_language_key = list(languages.keys())[list(languages.values()).index(c2.get())]

        # Get the input text
        input_text = txt1.get(1.0, END).strip()

        # Use GoogleTranslator to translate the text
        translated_text = GoogleTranslator(source=from_language_key, target=to_language_key).translate(input_text)

        # Display the translated text
        txt2.insert(1.0, translated_text)

    except Exception as e:
        messagebox.showerror("Translator", str(e))

# Text widgets for input and output
txt1 = Text(f1, font=('Arial', 12), width=28, height=8)
txt1.place(x=0, y=0)
txt2 = Text(f2, font=('Arial', 12), width=28, height=8)
txt2.place(x=0, y=0)

# Dropdown menus for language selection
c1 = ttk.Combobox(root, values=language_list, font=('Arial', 12), width=20)
c1.place(x=60, y=290)
c1.set("Select Source Language")  # Default text

c2 = ttk.Combobox(root, values=language_list, font=('Arial', 12), width=20)
c2.place(x=420, y=290)
c2.set("Select Target Language")  # Default text

# Translate button
btn = Button(root, text="Translate", font=('Arial', 15), bd=5, bg='#8B8B7A', fg='White', command=trans_late)
btn.place(x=290, y=320)

# Run the application
root.mainloop()
```
